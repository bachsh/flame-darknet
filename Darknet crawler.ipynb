```python
def get_tor_session():
    session = requests.session()
    # Tor uses the 9050 port as the default socks port
    session.proxies = {'http':  'socks5://127.0.0.1:9050',
                       'https': 'socks5://127.0.0.1:9050'}
    return session
```

```python
def download_product(url,forum_name,path):
    response = session.get(url,headers=headers)
    html = response.text
    id_num = hashlib.md5(url).hexdigest()
    data = {'html':html.encode('utf-8'),'url':url,'forum_name':forum_name}
    with open(os.path.join(path,str(id_num)+'.json'), 'wb') as outfile:
        json.dump(data,outfile)
        
```
